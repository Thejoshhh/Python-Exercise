```python
import random

valid_words = [
    "apple", "mango", "grape", "peach", "berry", "lemon", "plums", "cherry", "melon", "kiwi"
]


target_word = random.choice(valid_words)

MAX_GUESSES = 5


def give_feedback(guess, target):
    feedback = []
    
    for i, char in enumerate(guess):
        if char == target[i]:
            feedback.append(char.upper()) 
        elif char in target:
            feedback.append(char.lower()) 
        else:
            feedback.append('-')  
    
    return ''.join(feedback)


def play_wordle():
    print("Welcome to Wordle!")
    print(f"Try to guess the 5-letter word. You have {MAX_GUESSES} attempts.")
    
    for attempt in range(1, MAX_GUESSES + 1):
        print(f"\nAttempt {attempt}/{MAX_GUESSES}")
        guess = input("Enter your guess (5-letter word): ").lower()
        
 
        if len(guess) != 5:
            print("Please enter a 5-letter word.")
            continue
        if guess not in valid_words:
            print("Invalid word! Please try a valid 5-letter word.")
            continue
  
        feedback = give_feedback(guess, target_word)
        print("Feedback: ", feedback)
        
   
        if guess == target_word:
            print(f"Congratulations! You guessed the word: {target_word}")
            break
    else:
        print(f"\nGame Over! The correct word was: {target_word}")

play_wordle()
```

```python
import requests
import random

url = "https://gist.githubusercontent.com/dracos/dd0668f281e685bad51479e5acaadb93/raw/valid-wordle-words.txt"
response = requests.get(url)

# Split the content into a list of words
valid_words = response.text.splitlines()

# Select a random word as the target
target_word = random.choice(valid_words)

MAX_GUESSES = 5

def give_feedback(guess, target):
    feedback = ['-'] * len(guess)  # Initialize all as '-'
    target_copy = list(target)  # Create a mutable copy of the target to mark used letters

    # First pass: Check for correct letters in the correct positions
    for i, char in enumerate(guess):
        if char == target[i]:
            feedback[i] = char.upper()  # Correct letter, correct position
            target_copy[i] = None  # Mark this position as checked

    # Second pass: Check for correct letters in wrong positions
    for i, char in enumerate(guess):
        if feedback[i] == '-' and char in target_copy:
            feedback[i] = char.lower()  # Correct letter, wrong position
            target_copy[target_copy.index(char)] = None  # Mark this letter as used

    return ''.join(feedback)


def play_wordle():
    print("Welcome to Wordle!")
    print(f"Try to guess the 5-letter word. You have {MAX_GUESSES} attempts.")
    
    for attempt in range(1, MAX_GUESSES + 1):
        print(f"\nAttempt {attempt}/{MAX_GUESSES}")
        guess = input("Enter your guess (5-letter word): ").lower()
        
        if len(guess) != 5:
            print("Please enter a 5-letter word.")
            continue
        if guess not in valid_words:
            print("Invalid word! Please try a valid 5-letter word.")
            continue
        
        feedback = give_feedback(guess, target_word)
        print("Feedback: ", feedback)
        
        if guess == target_word:
            print(f"Congratulations! You guessed the word: {target_word}")
            break
    else:
        print(f"\nGame Over! The correct word was: {target_word}")

play_wordle()
```
